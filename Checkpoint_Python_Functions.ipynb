```python
def calculator(num1, num2):
  operator = input("Enter operator (+, -, *, /): ")
  match operator:
    case "+":
      return f"{num1} + {num2} = {num1 + num2}"
    case "-":
      return f"{num1} - {num2} = {num1 - num2}"
    case "*":
      return f"{num1} * {num2} = {num1 * num2}"
    case "/":
      if num2 == 0:
        return "Cannot divide by zero"
      return f"{num1} / {num2} = {num1 / num2}"
    case _:
      return "Invalid operator"

calculator(5, 2)
```
